```python
import random
from matplotlib import pyplot as plt

def montyhall(n, switch, interval=1):
    '''Simulates and plots the cumulative proportion
    of wins for the Monty Hall problem with n iterations.
    If switch is True the player will use the switch doors
    strategy, otherwise the player picks a random door
    and doens't switch
    '''
    doors = ['A', 'B', 'C']
    wins = 0
    y_axis = []
    x_axis = []
    
    if switch:
        for i in range(1, n+1):
            prize_door = random.choice(doors)
            player_door = random.choice(doors)
            host_door = random.choice(list(set(doors) - set(player_door) - set(prize_door)))
            (player_door,) = (set(doors) - set(player_door) - set(host_door)) # switches player door
            if player_door == prize_door:
                wins += 1
            if i % interval == 0:
                y_axis.append(wins/i)
                x_axis.append(i)
    else:
        for i in range(1, n+1):
            prize_door = random.choice(doors)
            player_door = random.choice(doors)
            if player_door == prize_door:
                wins += 1
            if i % interval == 0:
                y_axis.append(wins/i)
                x_axis.append(i)
            
    plt.plot(x_axis, y_axis)
    plt.xlabel('Number of games played')
    plt.ylabel('Winning proportion')
    if switch:
        plt.title("Switching door")
    else:
        plt.title("Not switching door")
    plt.show()
    
    
montyhall(10000,switch=True)
montyhall(10000,switch=False)
```
