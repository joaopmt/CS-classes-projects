```python
import random

def montyhall(n, switch):
    doors = ['A', 'B', 'C']
    wins = 0
    
    if switch:
        for i in range(1, n+1):
            prize_door = random.choice(doors)
            player_door = random.choice(doors)
            host_door = random.choice(list(set(doors) - set(player_door) - set(prize_door)))
            (player_door,) = (set(doors) - set(player_door) - set(host_door)) # switches player door
            if player_door == prize_door:
                wins += 1
            print(wins/i)
    else:
        for i in range(1, n+1):
            prize_door = random.choice(doors)
            player_door = random.choice(doors)
            if player_door == prize_door:
                wins += 1
            print(wins/i)
    

montyhall(5000,False)
```
