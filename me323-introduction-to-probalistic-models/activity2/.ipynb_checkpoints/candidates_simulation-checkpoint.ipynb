```python
import random
import statistics
from matplotlib import pyplot as plt

def simulate(n):
    y_axis = []
    population = 3000*[1] + 7000*[0]
    for i in range(1, 2001):
        sample = random.sample(population, k=n)
        prop_A = sum(sample)/len(sample)
        y_axis.append(prop_A)
    plt.figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.title('Graph for n = ' + str(n))
    plt.xlabel('Iteration')
    plt.ylabel('Proportion of votes on candidate A')
    plt.plot(y_axis)
    print('mean for n = ' + str(n) + ':', format(statistics.mean(y_axis), '.4f'))
    print('standard deviation for n = ' + str(n) + ':', format(statistics.stdev(y_axis), '.4f'))
    
    
simulate(10)
simulate(2000)
```

```python
Para a comparação, decidimos escolher um n = 10 e outro n = 2000.

Podemos observar claramente a diferença que temos usando esse intervalo grande para o n.
Com n = 10 vemos que a precisão é muito baixa e ocorre vezes em que a proporção de votos
no candidato A seja igual a 0.8 (quando esperávamos que fosse igual a 0.3). Já com n = 2000,
a precisão melhora muito e a diferença para o valor esperado (0.3) fica apenas na segunda
casa decimal, variando entre 0.26 e 0.34. Desses resultados podemos concluir que quanto
maior o nosso n, maior será a precisão e conseguiremos nos aproximar do valor esperado inicialmente.
```
